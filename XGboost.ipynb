## Dropping Irrelevant Columns

To improve model performance and reduce noise in the dataset, we dropped columns that were not useful for predicting the event_type. These included identifier fields, administrative metadata, tornado-specific details not relevant to most events, and text narratives not used in this model version.

```python
columns_to_drop = [
    'EVENT_ID', 'EPISODE_ID', 'ABSOLUTE_ROWNUMBER',
    'TOR_F_SCALE', 'TOR_LENGTH', 'TOR_WIDTH',
    'CZ_TYPE', 'CZ_FIPS', 'WFO', 'MAGNITUDE_TYPE',
    'INJURIES_INDIRECT', 'DEATHS_INDIRECT',
    'SOURCE', 'FLOOD_CAUSE',
    'BEGIN_LOCATION', 'BEGIN_RANGE', 'BEGIN_AZIMUTH',
    'END_LOCATION', 'END_DATE', 'END_TIME',
    'END_RANGE', 'END_AZIMUTH', 'END_LAT', 'END_LON',
    'EVENT_NARRATIVE', 'EPISODE_NARRATIVE',
    'CZ_TIMEZONE'
]

df.drop(columns=columns_to_drop, inplace=True)
```

## Feature Engineering to BeginDate


```python
# Convert date
df['BEGIN_DATE'] = pd.to_datetime(df['BEGIN_DATE'])

# Extract temporal features
df['YEAR'] = df['BEGIN_DATE'].dt.year
df['MONTH'] = df['BEGIN_DATE'].dt.month
df['WEEKDAY'] = df['BEGIN_DATE'].dt.weekday
df['HOUR'] = df['BEGIN_TIME'] // 100  # Converts 1630 to 16 (4PM)
```

```python
# Set a threshold for minimum frequency
threshold = 50

# Count class frequencies
event_counts = df['EVENT_TYPE'].value_counts()

# Identify rare event types
rare_event_types = event_counts[event_counts < threshold].index

# Replace them with 'Other' directly (as string, before encoding)
df['EVENT_TYPE_GROUPED'] = df['EVENT_TYPE'].apply(lambda x: 'Other' if x in rare_event_types else x)
```

```python
from sklearn.preprocessing import LabelEncoder

# Now encode EVENT_TYPE_GROUPED into 0-based class integers
le_event_grouped = LabelEncoder()
df['EVENT_TYPE_GROUPED_ENC'] = le_event_grouped.fit_transform(df['EVENT_TYPE_GROUPED'])
```

```python

# Convert numeric features
df['MAGNITUDE'] = pd.to_numeric(df['MAGNITUDE'], errors='coerce')
df['BEGIN_LAT'] = pd.to_numeric(df['BEGIN_LAT'], errors='coerce')
df['BEGIN_LON'] = pd.to_numeric(df['BEGIN_LON'], errors='coerce')

# Encode categorical input features
le_cz = LabelEncoder()
df['CZ_NAME_STR'] = le_cz.fit_transform(df['CZ_NAME_STR'])

le_state = LabelEncoder()
df['STATE_ABBR'] = le_state.fit_transform(df['STATE_ABBR'])
```

```python
df.dtypes
```

```python
# Drop the HOUR column from the DataFrame
df.drop(columns=['HOUR'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reconstruct label map from encoded values to string labels
label_map = dict(zip(df['EVENT_TYPE_GROUPED_ENC'], df['EVENT_TYPE_GROUPED']))

# Create readable label column (optional if not already added)
df['EVENT_TYPE_GROUPED_LABEL'] = df['EVENT_TYPE_GROUPED_ENC'].map(label_map)

# Plot the grouped distribution
plt.figure(figsize=(14, 6))
sns.countplot(
    x='EVENT_TYPE_GROUPED_LABEL',
    data=df,
    order=df['EVENT_TYPE_GROUPED_LABEL'].value_counts().index
)
plt.title('Event Type Distribution (Rare Types Grouped as "Other")')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## XGBoost Model


```python
# Drop label columns (human-readable ones) from features
X = df.drop(columns=['EVENT_TYPE', 'EVENT_TYPE_GROUPED', 'EVENT_TYPE_GROUPED_ENC', 'EVENT_TYPE_GROUPED_LABEL'])

# Define target
y = df['EVENT_TYPE_GROUPED_ENC']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.unique()),
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42,
    learning_rate=0.1,
    max_depth=6,
    n_estimators=100,
    subsample=0.8,
    colsample_bytree=0.8
)

xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = xgb_model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

importances = xgb_model.feature_importances_
features = X.columns
sorted_idx = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[sorted_idx][:15], y=features[sorted_idx][:15])
plt.title("Top 15 Feature Importances (XGBoost)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score

import pandas as pd
import numpy as np

metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'MAE', 'RMSE', 'R² Score'],
    'Value': [
        accuracy_score(y_test, y_pred),
        mean_absolute_error(y_test, y_pred),
        np.sqrt(mean_squared_error(y_test, y_pred)),
        r2_score(y_test, y_pred)
    ]
})

metrics_df['Value'] = metrics_df['Value'].round(4)
metrics_df
```

```python
y_pred_labels = le_event_grouped.inverse_transform(y_pred)
y_test_labels = le_event_grouped.inverse_transform(y_test)
```

```python
import pandas as pd

# Create prediction DataFrame (indexed like X_test)
prediction_df = pd.DataFrame({
    'True_Label': y_test_labels,
    'Predicted_Label': y_pred_labels
}, index=X_test.index)
```

```python
# Join key context features from X_test
context_cols = ['CZ_NAME_STR', 'MONTH', 'YEAR', 'STATE_ABBR']
prediction_df = prediction_df.join(X_test[context_cols])
```

```python
# Save to CSV
prediction_df.to_csv("xgboost_event_type_predictions.csv", index=False)
```
